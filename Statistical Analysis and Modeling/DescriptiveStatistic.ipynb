# Import required libraries:

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy import stats
import mplfinance as mpf
import seaborn as sns
```

# Load the Stock Price datasets:

Google:

```python
# Import the dataset
GOOG_df = pd.read_csv('GOOG.csv')

# Display the first few rows of the dataframe
print(GOOG_df.head())

# Convert the 'Date' column to datetime
GOOG_df['Date'] = pd.to_datetime(GOOG_df['Date'])

# Set the 'Date' column as the index
GOOG_df.set_index('Date', inplace=True)

print(GOOG_df.dtypes)
```

Microsoft:

```python
# Import the dataset
MSFT_df = pd.read_csv('MSFT.csv')

# Display the first few rows of the dataframe
print(MSFT_df.head())

# Convert the 'Date' column to datetime
MSFT_df['Date'] = pd.to_datetime(MSFT_df['Date'])

# Set the 'Date' column as the index
MSFT_df.set_index('Date', inplace=True)

print(MSFT_df.dtypes)
```

Apple:

```python
# Import the dataset
AAPL_df = pd.read_csv('AAPL.csv')

# Display the first few rows of the dataframe
print(AAPL_df.head())

# Convert the 'Date' column to datetime
AAPL_df['Date'] = pd.to_datetime(AAPL_df['Date'])

# Set the 'Date' column as the index
AAPL_df.set_index('Date', inplace=True)

print(AAPL_df.dtypes)
```

# Set a font for the plots:

```python
# Set the font family to Helvetica 
matplotlib.rcParams['font.family'] = 'Helvetica'
```

# Create a dataframe for Descriptive Statistics values:

Apple:

```python
AAPL_descriptive_stat = pd.DataFrame()

AAPL_descriptive_stat['Mean'] = AAPL_df.groupby(AAPL_df.index.year)['Close'].mean()
AAPL_descriptive_stat['Median'] = AAPL_df.groupby(AAPL_df.index.year)['Close'].median()
AAPL_descriptive_stat['Mode'] = AAPL_df.groupby(AAPL_df.index.year)['Close'].agg(lambda x: x.mode()[0])
AAPL_descriptive_stat['Variance'] = AAPL_df.groupby(AAPL_df.index.year)['Close'].var()
AAPL_descriptive_stat['Standard Deviation'] = AAPL_df.groupby(AAPL_df.index.year)['Close'].std()

print(AAPL_descriptive_stat.head())
```

Google:

```python
GOOG_descriptive_stat = pd.DataFrame()

GOOG_descriptive_stat['Mean'] = GOOG_df.groupby(GOOG_df.index.year)['Close'].mean()
GOOG_descriptive_stat['Median'] = GOOG_df.groupby(GOOG_df.index.year)['Close'].median()
GOOG_descriptive_stat['Mode'] = GOOG_df.groupby(GOOG_df.index.year)['Close'].agg(lambda x: x.mode()[0])
GOOG_descriptive_stat['Variance'] = GOOG_df.groupby(GOOG_df.index.year)['Close'].var()
GOOG_descriptive_stat['Standard Deviation'] = GOOG_df.groupby(GOOG_df.index.year)['Close'].std()

print(GOOG_descriptive_stat.head())
```

Microsoft:

```python
MSFT_descriptive_stat = pd.DataFrame()

MSFT_descriptive_stat['Mean'] = MSFT_df.groupby(MSFT_df.index.year)['Close'].mean()
MSFT_descriptive_stat['Median'] = MSFT_df.groupby(MSFT_df.index.year)['Close'].median()
MSFT_descriptive_stat['Mode'] = MSFT_df.groupby(MSFT_df.index.year)['Close'].agg(lambda x: x.mode()[0])
MSFT_descriptive_stat['Variance'] = MSFT_df.groupby(MSFT_df.index.year)['Close'].var()
MSFT_descriptive_stat['Standard Deviation'] = MSFT_df.groupby(MSFT_df.index.year)['Close'].std()

print(MSFT_descriptive_stat.head())
```

# Analyze Central Tendency Measures of AAPL's Closing Price:

Apple:

```python
plt.figure(figsize=(12, 6))

# Plot the mean, median, and mode
plt.plot(AAPL_descriptive_stat.index, AAPL_descriptive_stat['Mean'], label='Close Mean', color='#16c79a', marker='o')
plt.plot(AAPL_descriptive_stat.index, AAPL_descriptive_stat['Median'], label='Close Median', color='#2a0944', marker='o')
plt.plot(AAPL_descriptive_stat.index, AAPL_descriptive_stat['Mode'], label='Close Mode', color='#fb2576', marker='o')

# Add horizontal lines for the overall mean and median
plt.axhline(AAPL_df['Close'].mean(), color='#87CEFA', linestyle='--', label='Overall Close Mean')
plt.axhline(AAPL_df['Close'].median(), color='#FFA07A', linestyle='--', label='Overall Close Median')

# Plot the mode
plt.title('Mean, Median, and Mode of AAPL Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

# Plot the AAPL Monthly Candlestick Chart:

Apple:

```python
# Resample the data to monthly frequency
AAPL_monthly_df = AAPL_df.resample('ME').agg({
    'Open': 'first',
    'High': 'max',
    'Low': 'min',
    'Close': 'last',
    'Volume': 'sum'
})

# Define the style
custom_style = mpf.make_mpf_style(base_mpf_style='charles', 
                                  marketcolors=mpf.make_marketcolors(up='#3DC7A5', down='#E33F3E', edge='#4B0082', wick='#4B0082', volume='in'))

# Plot the candlestick chart with customizations
mpf.plot(AAPL_monthly_df, 
         type='candle', 
         style=custom_style, 
         title='Monthly Trend of AAPL Stock Price', 
         ylabel='Price',
         xlabel='Date',
         figsize=(12,6) ,
         volume=True) 
```

# Create a 6-Year AAPL Close Price Boxplot:

Apple:

```python
# Reset the index to extract the year 
AAPL_df_reset = AAPL_df.reset_index()

# Extract the year from the 'Date' column 
AAPL_df_reset['Year'] = AAPL_df_reset['Date'].dt.year

# Create a boxplot for the 'Close' column
plt.figure(figsize=(12, 6))
boxplot = sns.boxplot(x=AAPL_df_reset['Year'],y=AAPL_df_reset['Close'], color='#fb2576')
 
plt.title('A 6-Year Snapshot of AAPL Closing Prices', fontsize=18)
plt.ylabel('Close Price', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.grid(True)
plt.show()
```
